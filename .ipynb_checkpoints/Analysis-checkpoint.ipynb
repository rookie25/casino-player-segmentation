#  RFM ANALYSIS & PLAYER SEGMENTATION

**RFM Analysis** stands for:
- **R**ecency: How recently did the player last bet?
- **F**requency: How often does the player bet?
- **M**onetary: How much does the player wager?

This methodology segments players based on their behavioral patterns to identify high-value, engaged, and at-risk player groups.

##  Calculate Recency Score
Calculates how recently each player was active. More recent activity indicates higher engagement and retention likelihood.

```python
# ========================================
# CALCULATE RFM SCORES - COMPLETE
# ========================================
# Calculates Recency, Frequency, and Monetary scores with proper date parsing

print("CALCULATING RFM METRICS")
print("="*60)

# Step 1: Parse dates (format is M/D/YYYY)
print("\n1. Parsing dates...")
df['Pclstdate'] = pd.to_datetime(df['Pclstdate'], format='%m/%d/%Y', errors='coerce')
reference_date = pd.to_datetime('2007-02-28')

# Step 2: Calculate Recency (days since last activity)
df['Recency'] = (reference_date - df['Pclstdate']).dt.days
print(f"   Recency calculated: {df['Recency'].min():.0f} to {df['Recency'].max():.0f} days")
print(f"   Missing: {df['Recency'].isnull().sum()}")

# Step 3: Calculate Frequency (total active days)
df['Frequency'] = df['Pcsumday']
print(f"\n2. Frequency calculated: {df['Frequency'].min():.0f} to {df['Frequency'].max():.0f} days")
print(f"   Missing: {df['Frequency'].isnull().sum()}")

# Step 4: Calculate Monetary (total stakes)
df['Monetary'] = df['Pcsumstk']
print(f"\n3. Monetary calculated: ${df['Monetary'].min():.2f} to ${df['Monetary'].max():,.2f}")
print(f"   Missing: {df['Monetary'].isnull().sum()}")

# Step 5: Normalize to 0-100 scale
print("\n4. Normalizing scores...")

# Invert Recency (lower days = better = higher score)
max_recency = df['Recency'].max()
df['R_Score'] = ((max_recency - df['Recency']) / max_recency) * 100

# Scale Frequency
max_frequency = df['Frequency'].max()
df['F_Score'] = (df['Frequency'] / max_frequency) * 100

# Scale Monetary
max_monetary = df['Monetary'].max()
df['M_Score'] = (df['Monetary'] / max_monetary) * 100

print(f"   R_Score: {df['R_Score'].min():.2f} to {df['R_Score'].max():.2f}")
print(f"   F_Score: {df['F_Score'].min():.2f} to {df['F_Score'].max():.2f}")
print(f"   M_Score: {df['M_Score'].min():.2f} to {df['M_Score'].max():.2f}")

print("\n" + "="*60)
print("RFM SCORES CALCULATED SUCCESSFULLY!")
print("="*60)

print("\nSample of RFM scores:")
print(df[['UserID', 'Recency', 'Frequency', 'Monetary', 'R_Score', 'F_Score', 'M_Score']].head(10))
```

##  Calculate Monetary Score
Calculates the total amount wagered by each player. Higher monetary value indicates more valuable players to the casino.

```python
# ========================================
# CALCULATE MONETARY SCORE
# ========================================
# Uses total stakes as monetary measure

# Assigns monetary value (already exists as Pcsumstk)
df['Monetary'] = df['Pcsumstk']

print("MONETARY CALCULATION")
print("="*60)

print(f"\nMonetary Statistics:")
print(f"  Minimum: {df['Monetary'].min():.2f} Euros")
print(f"  Maximum: {df['Monetary'].max():,.2f} Euros")
print(f"  Average: {df['Monetary'].mean():,.2f} Euros")
print(f"  Median: {df['Monetary'].median():,.2f} Euros")

print("\nMonetary Distribution:")
print(df['Monetary'].describe())

print("\nPlayer Value Breakdown:")
low_value = len(df[df['Monetary'] < 1000])
med_value = len(df[(df['Monetary'] >= 1000) & (df['Monetary'] < 10000)])
high_value = len(df[df['Monetary'] >= 10000])

print(f"  Low Value (< 1,000 Euros): {low_value:,} players ({low_value/len(df)*100:.1f}%)")
print(f"  Medium Value (1,000-9,999 Euros): {med_value:,} players ({med_value/len(df)*100:.1f}%)")
print(f"  High Value (10,000+ Euros): {high_value:,} players ({high_value/len(df)*100:.1f}%)")

total_value = df['Monetary'].sum()
high_value_contribution = df[df['Monetary'] >= 10000]['Monetary'].sum()

print(f"\nTotal Casino Revenue: {total_value:,.2f} Euros")
print(f"Revenue from High Value Players: {high_value_contribution:,.2f} Euros ({high_value_contribution/total_value*100:.1f}%)")

print("\nSample of players with Monetary scores:")
print(df[['UserID', 'Pcsumstk', 'Pcsumday']].head(10))
```

```python
# ========================================
# FIX: RECALCULATE RECENCY
# ========================================
# Ensures Recency column exists before normalization

# Converts last active date to datetime if not already
if df['Pclstdate'].dtype == 'object':
    df['Pclstdate'] = pd.to_datetime(df['Pclstdate'], format='%d-%b-%Y', errors='coerce')

# Defines reference date
reference_date = pd.to_datetime('2007-02-28')

# Calculates recency
df['Recency'] = (reference_date - df['Pclstdate']).dt.days
# Assigns frequency and monetary if not already done
if 'Frequency' not in df.columns:
    df['Frequency'] = df['Pcsumday']
    
if 'Monetary' not in df.columns:
    df['Monetary'] = df['Pcsumstk']

print("RFM columns verified:")
print(f"  Recency: {'Recency' in df.columns}")
print(f"  Frequency: {'Frequency' in df.columns}")
print(f"  Monetary: {'Monetary' in df.columns}")
print("\nReady for normalization!")
```

## Normalize RFM Scores
Standardizes RFM scores to the same scale (0-100) to ensure equal weighting in the clustering algorithm. This prevents variables with larger ranges from dominating the segmentation.

```python
# ========================================
# COMPLETE RFM RECALCULATION
# ========================================
# Recalculates all RFM metrics from scratch

print("RECALCULATING RFM METRICS")
print("="*60)

# Step 1: Calculate Recency
print("\n1. Calculating Recency...")
df['Pclstdate'] = pd.to_datetime(df['Pclstdate'], format='%d-%b-%Y', errors='coerce')
reference_date = pd.to_datetime('2007-02-28')
df['Recency'] = (reference_date - df['Pclstdate']).dt.days

print(f"   Recency range: {df['Recency'].min():.0f} to {df['Recency'].max():.0f} days")
print(f"   Missing values: {df['Recency'].isnull().sum()}")

# Step 2: Calculate Frequency
print("\n2. Calculating Frequency...")
df['Frequency'] = df['Pcsumday']
print(f"   Frequency range: {df['Frequency'].min():.0f} to {df['Frequency'].max():.0f} days")
print(f"   Missing values: {df['Frequency'].isnull().sum()}")

# Step 3: Calculate Monetary
print("\n3. Calculating Monetary...")
df['Monetary'] = df['Pcsumstk']
print(f"   Monetary range: {df['Monetary'].min():.2f} to {df['Monetary'].max():,.2f} Euros")
print(f"   Missing values: {df['Monetary'].isnull().sum()}")

# Step 4: Normalize scores
print("\n4. Normalizing scores to 0-100...")

# Inverts and scales Recency
max_recency = df['Recency'].max()
df['R_Score'] = ((max_recency - df['Recency']) / max_recency) * 100

# Scales Frequency
max_frequency = df['Frequency'].max()
df['F_Score'] = (df['Frequency'] / max_frequency) * 100

# Scales Monetary
max_monetary = df['Monetary'].max()
df['M_Score'] = (df['Monetary'] / max_monetary) * 100

print(f"   R_Score range: {df['R_Score'].min():.2f} to {df['R_Score'].max():.2f}")
print(f"   F_Score range: {df['F_Score'].min():.2f} to {df['F_Score'].max():.2f}")
print(f"   M_Score range: {df['M_Score'].min():.2f} to {df['M_Score'].max():.2f}")

print("\n" + "="*60)
print("RFM CALCULATION COMPLETE")
print("="*60)
print(f"Missing R_Score: {df['R_Score'].isnull().sum()}")
print(f"Missing F_Score: {df['F_Score'].isnull().sum()}")
print(f"Missing M_Score: {df['M_Score'].isnull().sum()}")

print("\nSample of RFM scores:")
print(df[['UserID', 'R_Score', 'F_Score', 'M_Score']].head(10))
```

## Determine Optimal Number of Clusters
Uses the Elbow Method to identify the optimal number of player segments. Tests clustering with 2-8 segments and evaluates which provides the best separation.

```python
# ========================================
#  ELBOW METHOD FOR OPTIMAL CLUSTERS
# ========================================
# Determines the optimal number of player segments

from sklearn.cluster import KMeans

# Prepares data for clustering
X = df[['R_Score', 'F_Score', 'M_Score']].values

# Tests different numbers of clusters
inertias = []
K_range = range(2, 9)

print("Testing different numbers of clusters...")
print("="*60)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)
    print(f"K={k}: Inertia={kmeans.inertia_:.2f}")

# Creates elbow plot
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=12)
plt.title('Elbow Method: Optimal Number of Clusters', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks(K_range)

# Adds annotations
for i, k in enumerate(K_range):
    plt.annotate(f'{inertias[i]:.0f}', 
                xy=(k, inertias[i]), 
                xytext=(5, 5), 
                textcoords='offset points',
                fontsize=9)
plt.tight_layout()
plt.savefig('C:/Users/visha/OneDrive/Desktop/Casino Player Segmentation/reports/elbow_method.png', dpi=300, bbox_inches='tight')
print("\nElbow plot saved to reports/elbow_method.png")
plt.show()

print("\n" + "="*60)
print("RECOMMENDATION:")
print("="*60)
print("Look for the 'elbow' point where the curve bends.")
print("Typical optimal range: 4-6 clusters for player segmentation")
print("We will use K=5 for clear segment differentiation")
```

## K-Means Clustering
Applies K-means clustering with K=5 to segment players into distinct behavioral groups based on their RFM scores.

```python
# ========================================
# K-MEANS CLUSTERING (K=5)
# ========================================
# Applies K-means algorithm to create 5 player segments

from sklearn.cluster import KMeans

# Prepares data for clustering
X = df[['R_Score', 'F_Score', 'M_Score']].values

# Applies K-means with 5 clusters
print("Applying K-means clustering with K=5...")
print("="*60)

kmeans = KMeans(n_clusters=5, random_state=42, n_init=10, max_iter=300)
df['Segment'] = kmeans.fit_predict(X)

print("Clustering complete!")
print(f"\nCluster centers:")
print(kmeans.cluster_centers_)

print("\n" + "="*60)
print("SEGMENT DISTRIBUTION")
print("="*60)

segment_counts = df['Segment'].value_counts().sort_index()
for segment in range(5):
    count = segment_counts[segment]
    percentage = (count / len(df)) * 100
    print(f"Segment {segment}: {count:,} players ({percentage:.1f}%)")

print(f"\nTotal: {len(df):,} players")

print("\nSample of segmented players:")
print(df[['UserID', 'R_Score', 'F_Score', 'M_Score', 'Segment']].head(15))
```

## Segment Profiling
Analyzes each segment's characteristics by calculating average RFM scores, betting behavior, and player value to understand what defines each group.

```python
# ========================================
# SEGMENT PROFILING
# ========================================
# Analyzes characteristics of each player segment

print("SEGMENT PROFILING ANALYSIS")
print("="*60)

# Creates segment profile
segment_profile = df.groupby('Segment').agg({
    'R_Score': 'mean',
    'F_Score': 'mean',
    'M_Score': 'mean',
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'sum'],
    'UserID': 'count'
}).round(2)

segment_profile.columns = ['Avg_R_Score', 'Avg_F_Score', 'Avg_M_Score', 
                           'Avg_Recency_Days', 'Avg_Frequency_Days', 
                           'Avg_Monetary', 'Total_Revenue', 'Player_Count']

print("\nSegment Profiles:")
print(segment_profile)

print("\n" + "="*60)
print("SEGMENT CHARACTERISTICS SUMMARY")
print("="*60)

for segment in range(5):
    print(f"\n--- SEGMENT {segment} ---")
    profile = segment_profile.loc[segment]
    
    print(f"Size: {profile['Player_Count']:.0f} players")
    print(f"RFM Scores: R={profile['Avg_R_Score']:.1f}, F={profile['Avg_F_Score']:.1f}, M={profile['Avg_M_Score']:.1f}")
    print(f"Behavior: Last bet {profile['Avg_Recency_Days']:.0f} days ago, Active {profile['Avg_Frequency_Days']:.0f} days")
    print(f"Value: Avg ${profile['Avg_Monetary']:,.2f} per player")
    print(f"Total Revenue: ${profile['Total_Revenue']:,.2f}")
    
    pct_revenue = (profile['Total_Revenue'] / df['Monetary'].sum()) * 100
    print(f"Revenue Contribution: {pct_revenue:.1f}% of total")

print("\n" + "="*60)
```

## Segment Naming and Visualization
Assigns descriptive names to each segment based on their behavioral patterns and creates visualizations to compare segment characteristics.

```python
# ========================================
# SEGMENT NAMING
# ========================================
# Assigns descriptive names based on segment characteristics

print("ANALYZING SEGMENTS FOR NAMING...")
print("="*60)

# Analyzes each segment to determine appropriate names
segment_analysis = df.groupby('Segment').agg({
    'R_Score': 'mean',
    'F_Score': 'mean',
    'M_Score': 'mean',
    'Monetary': 'mean'
}).round(1)

print("\nSegment RFM Averages:")
print(segment_analysis)

# Creates segment name mapping based on RFM patterns
# High R, High F, High M = VIP/Champions
# Low R, High F, High M = At Risk High Value
# High R, Low F, Low M = New/Casual
# Low R, Low F, Low M = Lost/Dormant
# Medium across = Regular Players

segment_names = {}

for seg in range(5):
    r = segment_analysis.loc[seg, 'R_Score']
    f = segment_analysis.loc[seg, 'F_Score']
    m = segment_analysis.loc[seg, 'M_Score']
    
    if r > 60 and f > 60 and m > 60:
        name = 'VIP Champions'
    elif r > 60 and m > 30 and f > 30:
        name = 'Loyal High Value'
    elif r > 60 and f < 30 and m < 30:
        name = 'New Players'
    elif r < 40 and f > 40 and m > 40:
        name = 'At Risk High Value'
    elif r < 40 and f < 40 and m < 40:
        name = 'Dormant/Lost'
    elif m < 10 and f < 10:
        name = 'Low Value Casual'
    else:
        name = 'Regular Players'
    
    segment_names[seg] = name

# Applies names to dataframe
df['Segment_Name'] = df['Segment'].map(segment_names)

print("\n" + "="*60)
print("SEGMENT NAMES ASSIGNED")
print("="*60)

for seg in range(5):
    count = len(df[df['Segment'] == seg])
    pct = (count / len(df)) * 100
    avg_value = df[df['Segment'] == seg]['Monetary'].mean()
    print(f"Segment {seg}: {segment_names[seg]}")
    print(f"  Players: {count:,} ({pct:.1f}%)")
    print(f"  Avg Value: ${avg_value:,.2f}")
    print()

print("Sample with segment names:")
print(df[['UserID', 'R_Score', 'F_Score', 'M_Score', 'Segment', 'Segment_Name']].head(10))
```

```python
# ========================================
# FIX SEGMENT NAMES BASED ON ANALYSIS
# ========================================
# Assigns more accurate names based on actual characteristics

segment_names_corrected = {
    0: 'Low Value Dormant',
    1: 'Recent Low Spenders', 
    2: 'Inactive Casual',
    3: 'VIP High Rollers',
    4: 'New Casual Players'
}

df['Segment_Name'] = df['Segment'].map(segment_names_corrected)

print("CORRECTED SEGMENT NAMES")
print("="*60)

for seg in range(5):
    seg_data = df[df['Segment'] == seg]
    count = len(seg_data)
    pct = (count / len(df)) * 100
    avg_value = seg_data['Monetary'].mean()
    total_value = seg_data['Monetary'].sum()
    revenue_pct = (total_value / df['Monetary'].sum()) * 100
    
    print(f"\nSegment {seg}: {segment_names_corrected[seg]}")
    print(f"  Players: {count:,} ({pct:.1f}%)")
    print(f"  Avg RFM: R={seg_data['R_Score'].mean():.1f}, F={seg_data['F_Score'].mean():.1f}, M={seg_data['M_Score'].mean():.1f}")
    print(f"  Avg Value: ${avg_value:,.2f}")
    print(f"  Total Revenue: ${total_value:,.2f}")
    print(f"  Revenue Share: {revenue_pct:.1f}%")

print("\n" + "="*60)
print("KEY INSIGHT:")
print("="*60)
print(f"Segment 3 (VIP High Rollers): Only {len(df[df['Segment']==3])} players (2.9%)")
print(f"but generates ${df[df['Segment']==3]['Monetary'].sum():,.2f} in revenue!")
print(f"That's {(df[df['Segment']==3]['Monetary'].sum() / df['Monetary'].sum() * 100):.1f}% of total revenue!")
```

## Marketing Budget Allocation & ROI Analysis
Calculates optimal marketing budget distribution across segments based on lifetime value, segment size, and expected return on investment.

```python
# ========================================
# STEP 19: BUDGET ALLOCATION & ROI ANALYSIS
# ========================================
# Determines optimal marketing spend per segment

import numpy as np

print("MARKETING BUDGET ALLOCATION ANALYSIS")
print("="*60)

# Assumes total annual marketing budget
total_marketing_budget = 1000000  # $1 Million annual budget (adjust as needed)

print(f"\nTotal Annual Marketing Budget: ${total_marketing_budget:,}")

# Calculates segment metrics
segment_metrics = df.groupby('Segment_Name').agg({
    'UserID': 'count',
    'Monetary': ['mean', 'sum']
}).round(2)

segment_metrics.columns = ['Player_Count', 'Avg_LTV', 'Total_Revenue']
segment_metrics['Revenue_Share_Pct'] = (segment_metrics['Total_Revenue'] / df['Monetary'].sum() * 100).round(1)

# Defines investment strategy weights (based on segment priority)
# Higher weight = more budget allocation
investment_weights = {
    'VIP High Rollers': 40,           # Highest priority - 40% of budget
    'New Casual Players': 25,         # Growth opportunity - 25%
    'Recent Low Spenders': 20,        # Engagement boost - 20%
    'Inactive Casual': 10,            # Win-back - 10%
    'Low Value Dormant': 5            # Minimal - 5%
}

# Calculates budget allocation
segment_metrics['Investment_Weight'] = segment_metrics.index.map(investment_weights)
segment_metrics['Allocated_Budget'] = (segment_metrics['Investment_Weight'] / 100 * total_marketing_budget).round(2)
segment_metrics['Budget_Per_Player'] = (segment_metrics['Allocated_Budget'] / segment_metrics['Player_Count']).round(2)

print("\n" + "="*60)
print("BUDGET ALLOCATION BY SEGMENT")
print("="*60)
print(segment_metrics[['Player_Count', 'Revenue_Share_Pct', 'Investment_Weight', 
                       'Allocated_Budget', 'Budget_Per_Player']])

print("\n" + "="*60)
print("DETAILED BUDGET BREAKDOWN")
print("="*60)

for segment in segment_metrics.index:
    print(f"\n{segment}:")
    print(f"  Players: {segment_metrics.loc[segment, 'Player_Count']:,.0f}")
    print(f"  Revenue Share: {segment_metrics.loc[segment, 'Revenue_Share_Pct']:.1f}%")
    print(f"  Budget Allocation: ${segment_metrics.loc[segment, 'Allocated_Budget']:,.2f} ({investment_weights[segment]}%)")
    print(f"  Per Player Budget: ${segment_metrics.loc[segment, 'Budget_Per_Player']:,.2f}")
    print(f"  Current Avg LTV: ${segment_metrics.loc[segment, 'Avg_LTV']:,.2f}")

# ROI Analysis
print("\n" + "="*60)
print("EXPECTED ROI ANALYSIS")
print("="*60)

# Defines expected ROI multipliers (conservative estimates)
roi_multipliers = {
    'VIP High Rollers': 5.0,          # $1 spent returns $5 (retention)
    'New Casual Players': 3.0,        # $1 spent returns $3 (conversion)
    'Recent Low Spenders': 2.5,       # $1 spent returns $2.50
    'Inactive Casual': 1.5,           # $1 spent returns $1.50
    'Low Value Dormant': 1.2          # $1 spent returns $1.20
}

segment_metrics['Expected_ROI_Multiplier'] = segment_metrics.index.map(roi_multipliers)
segment_metrics['Expected_Return'] = (segment_metrics['Allocated_Budget'] * segment_metrics['Expected_ROI_Multiplier']).round(2)
segment_metrics['Net_Profit'] = (segment_metrics['Expected_Return'] - segment_metrics['Allocated_Budget']).round(2)

total_expected_return = segment_metrics['Expected_Return'].sum()
total_net_profit = segment_metrics['Net_Profit'].sum()
overall_roi = ((total_expected_return / total_marketing_budget) - 1) * 100

print("\nExpected Returns by Segment:")
for segment in segment_metrics.index:
    allocated = segment_metrics.loc[segment, 'Allocated_Budget']
    expected = segment_metrics.loc[segment, 'Expected_Return']
    profit = segment_metrics.loc[segment, 'Net_Profit']
    roi = segment_metrics.loc[segment, 'Expected_ROI_Multiplier']
    
    print(f"\n{segment}:")
    print(f"  Investment: ${allocated:,.2f}")
    print(f"  Expected Return: ${expected:,.2f} ({roi}x ROI)")
    print(f"  Net Profit: ${profit:,.2f}")

print("\n" + "="*60)
print("OVERALL MARKETING ROI")
print("="*60)
print(f"Total Investment: ${total_marketing_budget:,}")
print(f"Expected Return: ${total_expected_return:,.2f}")
print(f"Net Profit: ${total_net_profit:,.2f}")
print(f"Overall ROI: {overall_roi:.1f}%")

print("\n" + "="*60)
print("KEY RECOMMENDATIONS")
print("="*60)
print("\n1. PRIORITIZE VIP HIGH ROLLERS:")
print(f"   - Allocate ${segment_metrics.loc['VIP High Rollers', 'Allocated_Budget']:,.2f} (40% of budget)")
print(f"   - Highest ROI potential: 5x return")
print(f"   - Focus: Retention and increased engagement")

print("\n2. INVEST IN GROWTH (NEW CASUAL PLAYERS):")
print(f"   - Allocate ${segment_metrics.loc['New Casual Players', 'Allocated_Budget']:,.2f} (25% of budget)")
print(f"   - Conversion opportunity: 3x return")
print(f"   - Focus: Onboarding and frequency increase")

print("\n3. MINIMIZE LOW-VALUE SEGMENT SPEND:")
low_value_budget = segment_metrics.loc['Low Value Dormant', 'Allocated_Budget']
print(f"   - Only ${low_value_budget:,.2f} (5% of budget) on dormant players")
print(f"   - Low ROI: 1.2x return")
print(f"   - Focus: Automated campaigns only")

print("\n4. BALANCE ENGAGEMENT VS ACQUISITION:")
engagement_budget = (segment_metrics.loc['Recent Low Spenders', 'Allocated_Budget'] + 
                     segment_metrics.loc['Inactive Casual', 'Allocated_Budget'])
print(f"   - ${engagement_budget:,.2f} (30%) on re-engagement")
print(f"   - Complement with acquisition spending")
```
